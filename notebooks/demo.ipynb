## Bloom Filter Demo


Interactive demo walk through of the Bloomfilter Class

```python
import sys
from pathlib import Path

# Add project root to Python path
sys.path.append(str(Path.cwd().parent))
```

```python
from bloomfilter.bloom import BloomFilter

bf = BloomFilter(capacity=1000, error_rate=0.01)
print(bf)
```

```python
# Insert a few items
for word in ['apple', 'banana', 'cherry']:
    bf.add(word)

print('apple in bf  ->', 'apple' in bf)
print('grape in bf  ->', 'grape' in bf)
print('current fill ->', bf.fill_ratio())
```

## False Positive rate Measure

```python
bf = BloomFilter(capacity=10_000, error_rate=0.01)
for i in range(10_000):
    bf.add(f'in-{i}')

fp = sum(1 for i in range(10_000) if f'out-{i}' in bf)
print('measured FPR =', fp / 10_000)
```
